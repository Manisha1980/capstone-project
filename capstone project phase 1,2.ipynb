```python
import requests
from bs4 import BeautifulSoup
import csv

# Define the URL
url = 'https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1'

# Send a GET request and retrieve the HTML content
response = requests.get(url)

# Create a BeautifulSoup object to parse the HTML
soup = BeautifulSoup(response.content, 'html.parser')

# Extract the movie data
movies = soup.find_all('div', class_='lister-item')

# First CSV file: movie_details.csv
csv_file_1 = 'movie_details.csv'
csv_headers_1 = ['Sno', 'Movie Name', 'Director1', 'Director2', 'Duration', 'Year', 'Ratings', 'Metascore']

# Second CSV file: movie_info.csv
csv_file_2 = 'movie_info.csv'
csv_headers_2 = ['Movie Name', 'Star1', 'Star2', 'Star3', 'Star4', 'Votes', 'Genre1', 'Genre2', 'Genre3', 'Gross Collection', 'Popularity', 'Certification']

# Open the CSV files and write the headers
with open(csv_file_1, 'w', newline='', encoding='utf-8') as f1, open(csv_file_2, 'w', newline='', encoding='utf-8') as f2:
    writer1 = csv.writer(f1)
    writer1.writerow(csv_headers_1)

    writer2 = csv.writer(f2)
    writer2.writerow(csv_headers_2)

    # Extract and write the movie data to the CSV files
    for movie in movies:
        # Extract details for movie_details.csv
        sno = movie.find('span', class_='lister-item-index').text.strip('.')
        movie_name = movie.h3.a.text.strip()
        directors = movie.find_all('p', class_='text-muted')[1].find_all('a')
        director1 = directors[0].text.strip() if len(directors) >= 1 else ''
        director2 = directors[1].text.strip() if len(directors) >= 2 else ''
        duration = movie.find('span', class_='runtime').text.strip()
        year = movie.find('span', class_='lister-item-year').text.strip('()')
        ratings = movie.find('strong').text.strip()
        metascore = movie.find('span', class_='metascore').text.strip() if movie.find('span', class_='metascore') else ''

        writer1.writerow([sno, movie_name, director1, director2, duration, year, ratings, metascore])

        # Extract details for movie_info.csv
        stars = movie.find_all('p')[2].find_all('a')
        star1 = stars[0].text.strip() if len(stars) >= 1 else ''
        star2 = stars[1].text.strip() if len(stars) >= 2 else ''
        star3 = stars[2].text.strip() if len(stars) >= 3 else ''
        star4 = stars[3].text.strip() if len(stars) >= 4 else ''
        votes = movie.find('p', class_='sort-num_votes-visible').find_all('span')[-1].text.strip().replace(',', '')
        genres = movie.find('span', class_='genre').text.strip().split(',')
        genre1 = genres[0].strip() if len(genres) >= 1 else ''
        genre2 = genres[1].strip() if len(genres) >= 2 else ''
        genre3 = genres[2].strip() if len(genres) >= 3 else ''
        try:
            gross_collection = movie.find('p', class_='sort-num_gross_votes-visible').find_all('span')[-1].text.strip().replace(',', '')
        except AttributeError:
            gross_collection = ''  # Set empty string if gross collection is not available
        try:
            popularity = movie.find('div', class_='inline-block ratings-imdb-rating').strong['title'].split()[0]
        except (AttributeError, KeyError):
            popularity = ''  # Set empty string if popularity is not available
        certification = movie.find('span', class_='certificate').text.strip() if movie.find('span', class_='certificate') else ''

        writer2.writerow([movie_name, star1, star2, star3, star4, votes, genre1, genre2, genre3, gross_collection, popularity, certification])

# Print a success message
print("Data has been successfully stored in CSV files!")
```

```python
import sqlite3
import csv

# Connect to SQLite database
conn = sqlite3.connect('database.db')
cursor = conn.cursor()

# Create movie_details table
cursor.execute('''CREATE TABLE IF NOT EXISTS movie_details (
                    Sno INTEGER,
                    MovieName TEXT,
                    DirectorName TEXT,
                    Duration TEXT,
                    Year TEXT,
                    Ratings REAL,
                    Metascore TEXT
                )''')

# Create movie_info table
cursor.execute('''CREATE TABLE IF NOT EXISTS movie_info (
                    MovieName TEXT,
                    Stars TEXT,
                    Votes INTEGER,
                    Genre1 TEXT,
                    Genre2 TEXT,
                    Genre3 TEXT,
                    GrossCollection TEXT,
                    Popularity TEXT,
                    Certification TEXT
                )''')

# Import data from movie_details.csv
with open('movie_details.csv', 'r', newline='', encoding='utf-8') as f:
    csv_reader = csv.reader(f)
    next(csv_reader)  # Skip header row
    cursor.executemany("INSERT INTO movie_details VALUES (?, ?, ?, ?, ?, ?,?,?)", csv_reader)

# Import data from movie_info.csv
with open('movie_info.csv', 'r', newline='', encoding='utf-8') as f:
    csv_reader = csv.reader(f)
    next(csv_reader)  # Skip header row
    cursor.executemany("INSERT INTO movie_info VALUES (?, ?, ?, ?, ?, ?, ?, ?)", csv_reader)

# Commit the changes and close the connection
conn.commit()
conn.close()

print("Data imported into SQLite tables successfully!")
```
